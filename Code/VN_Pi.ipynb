```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def G(x):
    return np.sqrt(1-x**2)
```

```python
N=100000
A=0
Pi_i=[]
sqr_N=[]

for i in range(1,N+1):
    X=random.random()
    Y=random.random()
    if Y<G(X):
        A=A+1
    Pi_i.append(4*(A/i))
    sqr_N.append(np.sqrt(1/i)+np.pi)
    

Pi=4*(A/N)
print(Pi)

c_Pi=[np.pi]*N
plt.plot(Pi_i,)
plt.plot(c_Pi, 'r')
plt.plot(sqr_N)
plt.plot(-1*sqr_N)
plt.axis([0,N,2.8,4.2])
```
